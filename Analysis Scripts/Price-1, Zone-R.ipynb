# Summit County Sales Data sorted by Edgewood and Cascade Village (half mile boundaries) and Akron City

#### MAKE SURE THAT THE DATA FOR THE GRAPHS ARE FILTERED PROPERLY

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from dateutil.parser import parse
%matplotlib inline

# load the sales data, parsing the date
df = pd.read_csv('check_year.csv', parse_dates=[12], index_col=[0])
# get year from parsed date
df['year'] = df.sale_date.apply(lambda x: parse(x).year)

# load the list of parcels for cascade village, edgewood, and akron -- akron has a different schema
cv = pd.read_csv('parcels_halfmile_cv_buff.csv')
ed = pd.read_csv('edgewood_halfmi_parcels.csv')
ak = pd.read_csv('akron_residential_parcels.csv')
```

## Variable types from April

First pass:
* sale_type = 2 (land and building)
* sale_valid = 1 (valid sales)

Also run for:
* sale_type = 8 (foreclosure/sheriff's sales) OR
* sale_type = S (short sale) 

*Note: blank entries are not counted in groups*

## Let's take a closer look at sale_valid=='1' and sale_price>1 for residential parcels

```python
# filter df2 from df
df2 = df[(df.sale_price>1)]
```

```python
len(df2)
```

```python
def q2(x):
    return x.quantile(.25)
def q3(x):
    return x.quantile(.5)
def q4(x):
    return x.quantile(.75)

def get_by_year(sub, df, zone):
    sub = sub[sub['CLASS']==zone]
    df_sub = df[df.parcel_id.isin(sub.PARID)]
    df_sub['year'] = df_sub.sale_date.apply(lambda x: parse(x).year)
    sub_by_year = df_sub[['sale_price', 'year']].groupby('year').aggregate([np.median, len, np.average, q2, q3, q4])
    sub_by_year = sub_by_year[(sub_by_year.index>1997) & (sub_by_year.index<2015)]
    return sub_by_year

def get_by_year2(sub, df):
    df_sub = df[df.parcel_id.isin(sub.parcel_id)]
    df_sub['year'] = df_sub.sale_date.apply(lambda x: parse(x).year)
    sub_by_year = df_sub[['sale_price', 'year']].groupby('year').aggregate([np.median, len, np.average, q2, q3, q4])
    sub_by_year = sub_by_year[(sub_by_year.index>1997) & (sub_by_year.index<2015)]
    return sub_by_year

plt.style.use('fivethirtyeight')
```

```python
cv_by_year = get_by_year(cv, df2, 'R')
ed_by_year = get_by_year(ed, df2, 'R')
ak_by_year = get_by_year2(ak, df2)
```

### Let's make the plots year by year

###Average

```python
ax = plt.subplot(111)
plt.plot(ak_by_year.index, ak_by_year.sale_price['average'],  '.--', markersize=12,linewidth=2)
plt.plot(cv_by_year.index, cv_by_year.sale_price['average'], '.--', markersize=12,linewidth=2)
plt.plot(ed_by_year.index, ed_by_year.sale_price['average'], '.--', markersize=12,linewidth=2)
# plt.plot(sales_by_year.index, sales_by_year.sale_price['average'], '--', markersize=8)

plt.title('Average home price by year', y=1.08, x=.5)
plt.legend(['Akron','Cascade Village (half mile radius)','Edgewood (half mile radius)'], loc=(1.05,0.2))
plt.xlabel('Year')
plt.ylabel('Price ($)')
plt.xlim([1997,2016])
plt.ylim([-20000,320000])

ax.get_yaxis().set_major_formatter(plt.matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.annotate("Note: sale data are filtered by residential zone and sale_price > 1", (0,0), (0, -50), xycoords='axes fraction', textcoords='offset points', va='top')
# plt.savefig('average_cv.svg',bbox_inches='tight')
```

#Custom time ranges

All of this has been really excellent work. Some additional work: We're only going to use the half mile buffers (Cascade and Edgewood) and the city of akron. So you can put those three in a chart/table for available years.

Going to do a little something different with the x-axis of a second chart. We're going to create comparable periods of time relative to the housing redevelopment process.

Time periods: 
* Demo
* Reoccupancy
* Completion

Edgewood:
* Demo: 1998-2007
* Reoccupancy: 2008-2012
* Completion: 2013-today

Cascade:
* Demo: 1998-2004
* Demo/Reconstruction/Reoccupancy: 2005-2009
* Completion: 2010-today

Also, run what you can for commercial properties (same descriptives you've done, but just for the 3 geogs of interest). Try to get a sense of the descriptives for the "non valid" sales. Do what you can with sheriff's sales (transactions with sheriff's deeds, deed in lieu of foreclosure).

```python
ed2 = pd.DataFrame(index=['Demo','Reoccupancy','Completion'], columns=['total','median','average','years','middle_year'])
cv2 = pd.DataFrame(index=['Demo','Reoccupancy','Completion'], columns=['total','median','average','years','middle_year'])
```

```python
sub = ed[ed['CLASS']=='R']
df_sub = df2[df2.parcel_id.isin(sub.PARID)]
ed2['total'].ix['Demo'] = len(df_sub[(df_sub.year>=1998)&(df_sub.year<=2007)])
ed2['median'].ix['Demo'] = np.median(df_sub[(df_sub.year>=1998)&(df_sub.year<=2007)].sale_price)
ed2['average'].ix['Demo'] = np.mean(df_sub[(df_sub.year>=1998)&(df_sub.year<=2007)].sale_price)
ed2['years'].ix['Demo'] = 10
ed2['middle_year'].ix['Demo'] = 2002.5

ed2['total'].ix['Reoccupancy'] = len(df_sub[(df_sub.year>=2008)&(df_sub.year<=2012)])
ed2['median'].ix['Reoccupancy'] = np.median(df_sub[(df_sub.year>=2008)&(df_sub.year<=2012)].sale_price)
ed2['average'].ix['Reoccupancy'] = np.mean(df_sub[(df_sub.year>=2008)&(df_sub.year<=2012)].sale_price)
ed2['years'].ix['Reoccupancy'] = 5
ed2['middle_year'].ix['Reoccupancy'] = 2010

ed2['total'].ix['Completion'] = len(df_sub[(df_sub.year>=2013)&(df_sub.year<=2014)])
ed2['median'].ix['Completion'] = np.median(df_sub[(df_sub.year>=2013)&(df_sub.year<=2014)].sale_price)
ed2['average'].ix['Completion'] = np.mean(df_sub[(df_sub.year>=2013)&(df_sub.year<=2014)].sale_price)
ed2['years'].ix['Completion'] = 2
ed2['middle_year'].ix['Completion'] = 2013.5
```

```python
sub = cv[cv['CLASS']=='R']
df_sub = df2[df2.parcel_id.isin(sub.PARID)]
cv2['total'].ix['Demo'] = len(df_sub[(df_sub.year>=1998)&(df_sub.year<=2004)])
cv2['median'].ix['Demo'] = np.median(df_sub[(df_sub.year>=1998)&(df_sub.year<=2004)].sale_price)
cv2['average'].ix['Demo'] = np.mean(df_sub[(df_sub.year>=1998)&(df_sub.year<=2004)].sale_price)
cv2['years'].ix['Demo'] = 7
cv2['middle_year'].ix['Demo'] = 2001

cv2['total'].ix['Reoccupancy'] = len(df_sub[(df_sub.year>=2005)&(df_sub.year<=2009)])
cv2['median'].ix['Reoccupancy'] = np.median(df_sub[(df_sub.year>=2005)&(df_sub.year<=2009)].sale_price)
cv2['average'].ix['Reoccupancy'] = np.mean(df_sub[(df_sub.year>=2005)&(df_sub.year<=2009)].sale_price)
cv2['years'].ix['Reoccupancy'] = 5
cv2['middle_year'].ix['Reoccupancy'] = 2007

cv2['total'].ix['Completion'] = len(df_sub[(df_sub.year>=2010)&(df_sub.year<=2014)])
cv2['median'].ix['Completion'] = np.median(df_sub[(df_sub.year>=2010)&(df_sub.year<=2014)].sale_price)
cv2['average'].ix['Completion'] = np.mean(df_sub[(df_sub.year>=2010)&(df_sub.year<=2014)].sale_price)
cv2['years'].ix['Completion'] = 5
cv2['middle_year'].ix['Completion'] = 2012
```
