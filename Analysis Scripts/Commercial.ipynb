# Summit County Sales Data sorted by Edgewood and Cascade Village (half mile boundaries) and Akron City

**Always check the filters on the data before graphing it!**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# load the sales data, parsing the date
df = pd.read_csv('master.csv', index_col=[0])
```

```python
df.columns
```

## Variable types from April

First pass:
* sale_type = 2 (land and building)
* sale_valid = 1 (valid sales)

Also run for:
* sale_valid = 8 (foreclosure/sheriff's sales) OR
* sale_valid = S (short sale) 

And
* sales_valid in [1, 8, B, C, F, G, L, N, P, S, U, V, X, Z]
*Note: blank entries are not counted in groups*

```python
df2 = df[(df.sale_price>1) & (df.CLASS=='C') & (df.year > 1997) & (df.year < 2015)] # & (df.sale_valid.isin(['1','8','B','C','F','G','L','N','P','S','U','V','X','Z']))]
len(df2)
```

```python
def q2(x):
    return x.quantile(.25)
def q3(x):
    return x.quantile(.5)
def q4(x):
    return x.quantile(.75)

def get_by_year(df, var):
    df_sub = df[df[var]==1]
    return df_sub[['sale_price', 'year']].groupby('year').aggregate([np.median, len, np.average, q2, q3, q4])

plt.style.use('fivethirtyeight')
```

```python
cv_by_year = get_by_year(df2, 'cascade_village')
ed_by_year = get_by_year(df2, 'edgewood')
ak_by_year = get_by_year(df2, 'akron')
```

## Median Commercial Sale Price

```python
ax = plt.subplot(111)
plt.plot(ak_by_year.index, ak_by_year.sale_price['median'],  '.--', markersize=12,linewidth=2)
plt.plot(cv_by_year.index, cv_by_year.sale_price['median'], '.--', markersize=12,linewidth=2)
plt.plot(ed_by_year.index, ed_by_year.sale_price['median'], '.--', markersize=12,linewidth=2)

plt.title('Median commercial sale price by year', y=1.08, x=.5)
plt.legend(['Akron','Cascade Village (half mile radius)','Edgewood (half mile radius)'], loc=(1.05,0.2))
plt.xlabel('Year')
plt.ylabel('Value ($)')
plt.xlim([1997,2016])
plt.ylim([-100000,1200000])

ax.get_yaxis().set_major_formatter(plt.matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.annotate("Note: sale data filtered by zone (commercial), and sale_price > 1", (0,0), (0, -50), xycoords='axes fraction', textcoords='offset points', va='top')
# plt.savefig('average_cv.svg',bbox_inches='tight')
```

## Commercial Turnover by Year

```python
ax = plt.subplot(111)
plt.plot(ak_by_year.index, ak_by_year.sale_price['len']/len(set(df['parcel_id'][df['akron']==1])),  '.--', markersize=12,linewidth=2)
plt.plot(cv_by_year.index, cv_by_year.sale_price['len']/len(set(df['parcel_id'][df['cascade_village']==1])), '.--', markersize=12,linewidth=2)
plt.plot(ed_by_year.index, ed_by_year.sale_price['len']/len(set(df['parcel_id'][df['edgewood']==1])), '.--', markersize=12,linewidth=2)

plt.title('Commercial turnover by year', y=1.08, x=.5)
plt.legend(['Akron','Cascade Village (half mile radius)','Edgewood (half mile radius)'], loc=(1.05,0.2))
plt.xlabel('Year')
plt.ylabel('Sales per number of parcels')
plt.xlim([1997,2016])
plt.ylim([-0.001,0.022])

plt.annotate("Note: sale data filtered by zone (commercial), and sale_price > 1 \n         only parcels that show up in sales are counted", (0,0), (0, -50), xycoords='axes fraction', textcoords='offset points', va='top')
```

## Breakdown of count of commercial sale_valid code by geography

```python
def get_dist(df, var):
    df_sub = df[(df[var]==1) & (df['sale_valid'].isin(['1','8','B','C','F','G','L','N','P','S','U','V','X','Z']))]
    return df_sub[['sale_valid', 'year']].groupby('sale_valid').aggregate(len)

cv_dist = get_dist(df2, 'cascade_village')
ed_dist = get_dist(df2, 'edgewood')
ak_dist = get_dist(df2, 'akron')
```

```python
def fill_df(dist):
    for cat in ['1','8','B','C','F','G','L','N','P','S','U','V','X','Z']:
        if cat not in dist.index.values:
            dist = dist.append(pd.DataFrame([0],index = [cat], columns=['year']))
    return dist.sort_index()
```

```python
cv_dist = fill_df(cv_dist)
ed_dist = fill_df(ed_dist)
ak_dist = fill_df(ak_dist)
```

```python
def bar_plot(dist, name):
    y_pos = np.arange(len(dist.index))
    plt.bar(y_pos, dist.values, align='center', alpha=0.6)
    plt.xticks(y_pos, dist.index)
    plt.xlabel('sale_valid code')
    plt.title(name+' Commercial Sales')
    plt.gca().xaxis.grid(False)
    plt.annotate("Sales by category, 1998-2014", (0,0), (0, -50), xycoords='axes fraction', textcoords='offset points', va='top')
    plt.show()
```

```python
bar_plot(cv_dist,'Cascade Village')
```

```python
bar_plot(ed_dist, 'Edgewood')
```

## Let's look at sales filtered by validity code and geography

```python
def get_dist2(df, var):
    df_sub = df[(df[var]==1) & (df['sale_valid'].isin(['1','8','B','C','F','G','L','N','P','S','U','V','X','Z']))]
    agg = df_sub[['sale_valid', 'year']].groupby('year').aggregate(len)
    df_sub = df[(df[var]==1)]
    agg2 = df_sub[['sale_valid', 'year']].groupby('year').aggregate(len)
    
    return pd.merge(agg, agg2, how='inner', on=None, left_on=None, right_on=None,
      left_index=True, right_index=True, sort=True,
      suffixes=('_filter', '_all'), copy=True)

df2 =  df[(df.sale_price>1) & (df.CLASS=='C') & (df2.year > 1997) & (df2.year < 2015)] 
cv_dist2 = get_dist2(df2, 'cascade_village')
ed_dist2 = get_dist2(df2, 'edgewood')
ak_dist2 = get_dist2(df2, 'akron')
```

```python
def compare_plot(dist, name):
    ax = plt.subplot(111)
    plt.plot(dist.index, dist.sale_valid_filter, '.-', markersize=12,linewidth=2)
    plt.plot(dist.index, dist.sale_valid_all, '.-', markersize=12,linewidth=2)

    plt.title('Commercial sales in '+name, y=1.08, x=.5)
    plt.legend(['All sales','Filtered sales'], loc=(1.05,0.2))
    plt.xlabel('Year')
    plt.xlim([1997,2016])

    ax.get_yaxis().set_major_formatter(plt.matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
    plt.annotate("Note: Filtered sales means sale_valid in [1, 8, B, C, F, G, L, N, P, S, U, V, X, Z],\n          commercial zone, and sale_price > 1", (0,0), (0, -50), xycoords='axes fraction', textcoords='offset points', va='top')
    # plt.savefig('average_cv.svg',bbox_inches='tight')
```

#Custom time ranges

Going to do a little something different with the x-axis of a second chart. We're going to create comparable periods of time relative to the housing redevelopment process.

Time periods: 
* Demo
* Reoccupancy
* Completion

Edgewood:
* Demo: 1998-2007
* Reoccupancy: 2008-2012
* Completion: 2013-today

Cascade:
* Demo: 1998-2004
* Demo/Reconstruction/Reoccupancy: 2005-2009
* Completion: 2010-today

Also, run what you can for commercial properties (same descriptives you've done, but just for the 3 geogs of interest). Try to get a sense of the descriptives for the "non valid" sales. Do what you can with sheriff's sales (transactions with sheriff's deeds, deed in lieu of foreclosure).

```python
ed2 = pd.DataFrame(index=['Demo','Reoccupancy','Completion'], columns=['total','median','average','years','middle_year'])
cv2 = pd.DataFrame(index=['Demo','Reoccupancy','Completion'], columns=['total','median','average','years','middle_year'])
```

```python
df_sub = df[(df.edgewood==1) & (df.CLASS == 'C') & (df.sale_price>1)]
ed2['total'].ix['Demo'] = len(df_sub[(df_sub.year>=1998)&(df_sub.year<=2007)])
ed2['median'].ix['Demo'] = np.median(df_sub[(df_sub.year>=1998)&(df_sub.year<=2007)].sale_price)
ed2['average'].ix['Demo'] = np.mean(df_sub[(df_sub.year>=1998)&(df_sub.year<=2007)].sale_price)
ed2['years'].ix['Demo'] = 10
ed2['middle_year'].ix['Demo'] = 2002.5

ed2['total'].ix['Reoccupancy'] = len(df_sub[(df_sub.year>=2008)&(df_sub.year<=2012)])
ed2['median'].ix['Reoccupancy'] = np.median(df_sub[(df_sub.year>=2008)&(df_sub.year<=2012)].sale_price)
ed2['average'].ix['Reoccupancy'] = np.mean(df_sub[(df_sub.year>=2008)&(df_sub.year<=2012)].sale_price)
ed2['years'].ix['Reoccupancy'] = 5
ed2['middle_year'].ix['Reoccupancy'] = 2010

ed2['total'].ix['Completion'] = len(df_sub[(df_sub.year>=2013)&(df_sub.year<=2014)])
ed2['median'].ix['Completion'] = np.median(df_sub[(df_sub.year>=2013)&(df_sub.year<=2014)].sale_price)
ed2['average'].ix['Completion'] = np.mean(df_sub[(df_sub.year>=2013)&(df_sub.year<=2014)].sale_price)
ed2['years'].ix['Completion'] = 2
ed2['middle_year'].ix['Completion'] = 2013.5
```

```python
df_sub = df[(df.cascade_village==1) & (df.CLASS == "C") & (df.sale_price>1)]
cv2['total'].ix['Demo'] = len(df_sub[(df_sub.year>=1998)&(df_sub.year<=2004)])
cv2['median'].ix['Demo'] = np.median(df_sub[(df_sub.year>=1998)&(df_sub.year<=2004)].sale_price)
cv2['average'].ix['Demo'] = np.mean(df_sub[(df_sub.year>=1998)&(df_sub.year<=2004)].sale_price)
cv2['years'].ix['Demo'] = 7
cv2['middle_year'].ix['Demo'] = 2001

cv2['total'].ix['Reoccupancy'] = len(df_sub[(df_sub.year>=2005)&(df_sub.year<=2009)])
cv2['median'].ix['Reoccupancy'] = np.median(df_sub[(df_sub.year>=2005)&(df_sub.year<=2009)].sale_price)
cv2['average'].ix['Reoccupancy'] = np.mean(df_sub[(df_sub.year>=2005)&(df_sub.year<=2009)].sale_price)
cv2['years'].ix['Reoccupancy'] = 5
cv2['middle_year'].ix['Reoccupancy'] = 2007

cv2['total'].ix['Completion'] = len(df_sub[(df_sub.year>=2010)&(df_sub.year<=2014)])
cv2['median'].ix['Completion'] = np.median(df_sub[(df_sub.year>=2010)&(df_sub.year<=2014)].sale_price)
cv2['average'].ix['Completion'] = np.mean(df_sub[(df_sub.year>=2010)&(df_sub.year<=2014)].sale_price)
cv2['years'].ix['Completion'] = 5
cv2['middle_year'].ix['Completion'] = 2012
```
